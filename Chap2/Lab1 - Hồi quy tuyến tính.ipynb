```python
# Hồi quy tuyến tính
import numpy as np # solve lib
import matplotlib.pyplot as plt #draw lib
```

```python
 # weight data
x = np.array([48, 49, 75, 65, 80, 45])

#tall data
y = np.array([155, 157, 175, 175, 185, 158])

print(x)
print(y)
```

```python
plt.scatter(x, y)
```

```python
# find w1 and w0
# use a lib
from sklearn.linear_model import LinearRegression # Hồi quy tuyến tính
```

```python
reg = LinearRegression() # create an instance

# draw estimate line
reg.fit(x.reshape(-1, 1), y) # 1: column
```

```python
reg.coef_[0] #find w1
```

```python
reg.intercept_ #find w0
```

```python
# draw to Oxy
y_pre = reg.coef_[0] * x + reg.intercept_ 
print(y_pre)
```

```python
plt.scatter(x, y, c = 'blue')
plt.scatter(x, y_pre, c = 'red') # draw points
plt.plot(x, y_pre, c = 'red') # draw a line
```

```python
# enter a data and test
x_test = 54

y_pre_test = reg.coef_[0] * x_test + reg.intercept_
print(y_pre_test)

plt.scatter(x, y, c = 'blue')
plt.scatter(x, y_pre, c = 'red') # draw points
plt.plot(x, y_pre, c = 'red') # draw a line
plt.scatter(x_test, y_pre_test, c = 'green')
```
