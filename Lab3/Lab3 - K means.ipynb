```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs # hàm dùng tạo dữ liệu ngẫu nhiên
from sklearn.cluster import KMeans # hàm đã được implement cho thuật toán K means
```

```python
x, _ = make_blobs(n_samples=200, n_features=2, cluster_std=3, random_state=50)

x.shape
```

```python
x1 = x[:, 0]
x2 = x[:, 1]

plt.scatter(x1, x2)
```

```python
kmeans = KMeans(n_clusters=3, random_state=13)

kmeans = kmeans.fit(x)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
color_dict = {0: 'r', 1: 'g', 2: 'b'}
colors = [color_dict[i] for i in kmeans.labels_] # gán màu cho từng điểm dữ liệu
```

```python
plt.scatter(x1, x2, color = colors)
```

```python
# vẽ centroid
plt.scatter(x1, x2, color = colors)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = 'yellow')
```

```python
# tạo 1 điểm dữ liệu và xem nó thuộc cụm nào
test_point = np.array([13, 0])

test_point_label = kmeans.predict(test_point.reshape(-1, 2))

print(test_point_label)
```

```python
plt.scatter(x1, x2, color = colors)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = 'yellow')
plt.scatter(test_point[0], test_point[1], color = color_dict[test_point_label[-1]], marker= '*', s=100)
```
